```python
import jieba
import wordcloud
datafile="C:/Users/CCEC/Desktop/实验7数据（新版）/实验7数据/introduce.txt"
dictfile="C:/Users/CCEC/Desktop/实验7数据（新版）/实验7数据/mydic.txt"
maskimg="C:/Users/CCEC/Desktop/实验7数据（新版）/实验7数据/tree.jpg"
f=open(datafile,encoding="GBK")
s=f.read()
f.close()
for i in "'\n'。，“”）（":
    s=s.replace(i,"")
s
```

```python
jieba.load_userdict(dictfile)
words=jieba.cut(s)
words=list(words)
wordstr=" ".join(words)
wordstr
```

```python
mask=np.array(Image.open(maskimg))
plt.imshow(mask)
plt.axis('off')
plt.show()
```

```python
wc=wordcloud.WordCloud(background_color='white',width=800,height=600,
                      max_words=300,max_font_size=80,mask=mask,
                      font_path='C:/Users/CCEC/Desktop/实验7数据（新版）/实验7数据/SIMLI.TTF',random_state=100,scale=1.5)
wc.generate(wordstr)
plt.imshow(wc)
plt.axis('off')
plt.show()
```
