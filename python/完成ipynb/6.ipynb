```python
df["一级lu1980"]=[x//10 if x<100 else x//100 for x in df["lu1980"]]
df["一级lu2010"]=[x//10 for x in df["lu2010"]]
df.head()
```

```python
import numpy as np
luname=["耕地","林地","草地","水域","建设用地","未利用地"]
df["area"]=(df["area"]/1000000).round(2)
```

```python
pd.pivot_table(df,values="area",columns="一级lu2010",index="city",aggfunc=np.sum)
```

```python
adminlu=pd.crosstab(values=df["area"],index=df["city"],columns=df["一级lu2010"],aggfunc=np.sum,margins=True,margins_name="总体")
luname=["耕地","林地","草地","水域","建设用地","未利用地","总体"]
adminlu.columns=luname
adminlu
```

```python
columns=["耕地","林地","草地","水域","建设用地","未利用地","合计"]
admin=df[["city","city_code"]].drop_duplicates()
for name,code in admin.values:
    dfcountry=df.loc[df["city"]==name]
    countrylu=pd.crosstab(values=dfcountry["area"],index=dfcountry["district"],columns=dfcountry["一级lu2010"],aggfunc=np.sum,margins=True,margins_name="总体")
    #dfadminlu.columns=luname
    countrylu.to_excel(filepath+"lustat1/"+str(code)+"_"+name+"各县市土地利用面积统计表.xls")
    print(name,"数据处理完毕")
```

```python
heblu=df[df["city"]=="哈尔滨市"]
hebchg=pd.crosstab(values=heblu["area"],index=heblu["一级lu1980"],columns=heblu["一级lu2010"],aggfunc=np.sum,margins=True,margins_name="总体")
hebchg["lu1980"]=luname
hebchg.set_index("lu1980",inplace=True)
hebchg.columns=luname
hebchg
```

```python
area1980=hebchg["总体"]
area1980.name="1980"
area1980.index.name="lu"
area2010=hebchg.loc["总体"]
area2010.name="2010"
area2010.index.name="年份"
pd.DataFrame([area1980,area2010])
```
